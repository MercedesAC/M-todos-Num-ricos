```python
 # Definicion de la funcion
fx= lambda x: np.sin(x)/(x)
# intervalo de la integral
a = 0.0000001
b = 2
# tramos o divisiones 
tramos = 10   
i = 0
```

```python
 # PROCEDIMIENTO DE CALCULO
h = (b-a)/tramos # base cuya es igual al intervalo divido por los tramos  
muestras = tramos + 1 # cantidad de muestras
suma = 0 # variable donde se almacenara la suma de las areas
xi = a 
Atrapecio = fx(xi) ## valor del area del trapecio 
suma = suma + Atrapecio
i = i+1

while i<(tramos -1): # bucle que nos permite calcular la are cuya recoge cada uno de los tramos 
    Atrapecio = 2*(fx(xi)) ## valor del area del trapecio 
    suma = suma + Atrapecio
    xi = xi + h 
    i = i + 1
 
xi = xi + h
Atrapecio = fx(xi) ## valor del area del trapecio 
suma = suma + Atrapecio
suma = suma * h/2
integral = suma
          
xi = np.linspace(a, b, muestras) # distribuicion de los puntos entre a y b con la cantidad de muestras
fi = fx(xi) #calculo de puntos dentro de la funcion


muestraslinea = muestras*10 # muestras de las lineas
xk = np.linspace(a, b, muestraslinea) # distribucion de puntos entre a y b 
                                      #con la cantidad necesaria de muestras de las lineas 
fk = fx(xk) # evaluacion de los puntos en la posicion "k"
          
```

```python
# Resultados 
print ('Tramos: ', tramos) # impresion de la cantidad de tramos usados  para la integracion
print ('Integral: ', integral) # impresion del resultado de la integral

 # grafica de los puntos marcados
plt.plot(xi,fi, 'ro') 
plt.plot(xk,fk)
plt.show()
```

```python
# Generacion de trapecios 
plt. fill_between(xi, 0, fi, color='g') # relleno entre la funcion que pasa por los puntos donde se realizo el muestreo 
for i in range (0, muestras, 2): # bucle que recoge cada punto del muestreo 
    plt.axvline(xi[i], color = 'w') # lineas que ensenan los limites del trapecio
plt.show ()
```
