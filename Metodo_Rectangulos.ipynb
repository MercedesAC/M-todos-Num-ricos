## EJERCICIO 1

```python
# importacion de librerias 
import numpy as np # Libreria que nos permitira43 revisar los calculos 
import matplotlib.pyplot as plt # Libreria que nos permitira revisar las graficas -
from IPython.display import Image
```

```python
Image('ejercicio.png', width = 200, height = 200)
```

```python
# creamos el metodp retangulo paracalcular el area en el area determinado 
def f(x):
     return 1/(1 + x**2) 
n = 8 # subdivisiones del intervalo / numero de interaciones 
suma = 0 # variable cuya se va atribuir la suma de las areas de los intervalos 
# planteamiento del intervalo de nuestra integral
a = 0
b = 1

# Definimos el diferencial/ base de cada retangulo
Dx = (b - a) / n

# Bucle para calculo de cada intervalo
for i in range(1, n+1):
    altura = f(a) # altura del retangulo
    Area = Dx * altura #  Area del retangulo
    suma = suma + Area 
    a = a + Dx # encontramos un nuevo a para el siguiente calculo

print (suma) # resultado de la suma total 
print ('El área aproximada es de' ,abs(suma)) # resultado del valor del area aproximada
```

```python
x = np.linspace(-2, 6, 101)
# cordenadas de la grafica
plt.plot(x, 1/(1 + x**2)) #comportamiento de la funcion
plt.plot(x, np.linspace(0, 0, 101)) # una linea para definir el eje x
plt.grid()
plt.show()
```
